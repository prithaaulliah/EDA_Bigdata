```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Dataset
data = pd.read_csv("courses.csv")
```

```python
#Print 5 data awal
data.head(5)
```

```python
#Print 5 data akhir
data.tail(5)
```

```python
# Melihat informasi dasar tentang dataset
data.info()
```

```python
data.describe()
```

```python
# Melihat jumlah missing values pada setiap kolom
print("\nJumlah Missing Values pada Setiap Kolom:")
print(data.isnull().sum())
```

```python
grouped_stats = data.groupby('code_module')['module_presentation_length'].describe() #Gabungkan code_module dengan module_presentation_length
print("Statistik Deskriptif Berdasarkan Kelompok Code Module:\n", grouped_stats)
```

```python
data_count = data.groupby(['code_module', 'code_presentation']).size().reset_index(name='count')
print("Jumlah Data per Code Module dan Code Presentation:\n", data_count)
```

```python
#Melakukan Plotting menggunakan diagram batang
plt.figure(figsize=(8, 6))
sns.histplot(data['module_presentation_length'], bins=10, kde=True)
plt.xlabel('Module Presentation Length')
plt.ylabel('Frequency')
plt.title('Distribution of Module Presentation Length')
plt.show()
```

```python
#Melakukan plotting menggunakan box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='code_module', y='module_presentation_length', data=data)
plt.xlabel('Code Module')
plt.ylabel('Module Presentation Length')
plt.title('Module Presentation Length by Code Module')
plt.xticks(rotation=45)
plt.show()
```

```python
#Melakukan plotting menggunakan scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='code_module', y='code_presentation', data=data, hue='module_presentation_length')
plt.xlabel('Code Module')
plt.ylabel('Code Presentation')
plt.title('Scatter Plot of Code Module and Code Presentation')
plt.xticks(rotation=45)
plt.show()
```

```python
#Melakukan plotting data code_module menggunakan histrogram
plt.figure(figsize=(10, 6))
sns.countplot(x='code_module', data=data)
plt.xlabel('Code Module')
plt.ylabel('Count')
plt.title('Count of Data per Code Module')
plt.xticks(rotation=45)
plt.show()
```

```python
#Melakukan plotting code_presentation menggunakan histogram
plt.figure(figsize=(10, 6))
sns.countplot(x='code_presentation', data=data)
plt.xlabel('Code Presentation')
plt.ylabel('Count')
plt.title('Count of Data per Code Presentation')
plt.xticks(rotation=45)
plt.show()
```
