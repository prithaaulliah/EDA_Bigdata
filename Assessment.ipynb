```python
# Visualisasi Box Plot
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.boxplot(df['id_assessment'])
plt.title('Box Plot id_assessment')

plt.subplot(1, 3, 2)
sns.boxplot(df['date'])
plt.title('Box Plot date')

plt.subplot(1, 3, 3)
sns.boxplot(df['weight'])
plt.title('Box Plot weight')

plt.tight_layout()
plt.show()

```

```python
# Visualisasi Distribusi Data dengan Histogram
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.histplot(df['id_assessment'], bins=20, kde=True)
plt.title('Distribusi id_assessment')

plt.subplot(1, 3, 2)
sns.histplot(df['date'], bins=20, kde=True)
plt.title('Distribusi date')

plt.subplot(1, 3, 3)
sns.histplot(df['weight'], bins=20, kde=True)
plt.title('Distribusi weight')

plt.tight_layout()
plt.show()
```

```python
# Menghitung matriks korelasi
correlation_matrix = df.corr()

# Membuat heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriks Korelasi')
plt.show()
```

```python
# Membuat scatter plot untuk memeriksa interaksi
plt.figure(figsize=(8, 6))
sns.boxplot(x='assessment_type', y='weight', data=df)
plt.title('Hubungan Assessment Type dan Weight')
plt.xlabel('Assessment Type')
plt.ylabel('Weight')
plt.show()
```

```python
# Membuat scatter plot untuk memeriksa interaksi
plt.figure(figsize=(12, 6))
sns.boxplot(x='code_module', y='weight', data=df)
plt.title('Hubungan Code Module dan Weight')
plt.xlabel('Code Module')
plt.ylabel('Weight')
plt.show()
```
